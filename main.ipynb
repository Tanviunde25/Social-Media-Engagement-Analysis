```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("social_media_engagement1.csv")

avg = df[['likes','comments','shares']].mean()

print("Average Engagement:")
print(avg)

plt.figure(figsize=(8,5))

plt.plot(avg.index, avg.values, marker='o')

plt.xlabel("Engagement Type")
plt.ylabel("Average Count")
plt.title("Average Social Media Engagement")

plt.show()
```
